```python
import pandas as pd
import numpy as np
df_abalone = pd.read_csv('abalone.data', header=None)
df_abalone.columns = ["Sex", "Length", "Diameter", "Height", "Whole weight", "Shucked weight", "Viscera weight", "Shell weight", "Rings"]
df_abalone
```

```python
data = df_abalone.sample(frac=0.9, random_state=786)
data_unseen = df_abalone.drop(data.index)

data.reset_index(drop=True, inplace=True)
data_unseen.reset_index(drop=True, inplace=True)

print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

```python
exp_mclf101 = setup(data = data, target = 'Sex', session_id=123) 
```

```python
best = compare_models()
```

```python
dectree = create_model('dt')
```

```python
print(dectree)
```

```python
knn = create_model('knn')
```

```python
lr = create_model('lr')
```

```python
tuned_dectree = tune_model(dectree)
```

```python
print(tuned_dectree)
```

```python
tuned_knn = tune_model(knn, custom_grid = {'n_neighbors' : np.arange(0,50,1)})
```

```python
tuned_lr = tune_model(lr)
```

```python
plot_model(tuned_lr, plot = 'confusion_matrix')
```

```python
plot_model(tuned_lr, plot = 'class_report')
```

```python
plot_model(tuned_lr, plot='boundary')
```

```python
plot_model(tuned_lr, plot = 'error')
```

```python
evaluate_model(tuned_lr)
```

```python
predict_model(tuned_lr);
```

```python
final_lr = finalize_model(tuned_lr)
```

```python
print(final_lr)
```

```python
unseen_predictions = predict_model(final_lr, data=data_unseen)
unseen_predictions.head()
```

```python
save_model(final_lr,'Final LR Model')
```

```python
saved_final_lr = load_model('Final LR Model')
```
